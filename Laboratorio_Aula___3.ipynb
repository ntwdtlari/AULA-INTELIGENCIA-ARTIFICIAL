# 2. Armazenando interações - Listas, dicionários e funções

```python
print('Bem-vindo ao ChatBot da Univille AI ! (Digite x se você quiser sair!)\n') #BOAS VINDAS
mensagens = [] #LISTA DE ARMAZENAMENTO DA CONVERSA
while True:
  pergunta = input('Usuário: ')
  if pergunta.lower() == 'x':
    break


  mensagens.append({'role': 'user', 'content': pergunta}) #CONTEÚDO DE INTERAÇÃO DO USUÁRIO
  resposta = 'Resposta do bot' #INTERAÇÃO DO BOT
  mensagens.append({'role': 'assistant', 'content': resposta}) #CONTEÚDO GERADO PELO ASSISTENTE VIRTUAL
  print(f'Bot: {resposta}')

print('\nMuito obrigado por utilizar o Univille AI! O histórico completo da conversa foi:')
print(mensagens)
```

```python
import os
from langchain_groq import ChatGroq
from langchain.prompts import ChatPromptTemplate

#Configure sua API KEY aqui
api_key = 'SUA_API_KEY_AQUI'
os.environ['GROQ_API_KEY'] = api_key

#Modelo
chat = ChatGroq(model='llama-3.3-70b-versatile')

#Escolha do tipo de bot
print('=== UNIBOT PERSONALIZADO ===')
print('Escolha o tipo de assistente:')
print('1 - Professor')
print('2 - Programador')
print('3 - Motivacional')
print('4 - Gamer')
print('5 - Nutricionista')

opcao = input('Digite o número desejado: ')

#Definição do comportamento (contexto)
def definir_contexto(opcao):
    if opcao == '1':
        return 'Você é um professor da Univille que explica tudo de forma simples e didática'
    elif opcao == '2':
        return 'Você é um programador especialista em Python e sempre responde com código e explicações'
    elif opcao == '3':
        return 'Você é um coach motivacional que sempre incentiva o usuário com frases inspiradoras'
    elif opcao == '4':
        return 'Você é um gamer experiente e fala de forma descontraída, usando linguagem de jogos'
    elif opcao == '5':
        return 'Você é um nutricionista que sugere alimentação saudável e equilibrada'
    else:
        return 'Você é um assistente virtual amigável'

#Função do bot
def resposta_bot(mensagens, contexto):
    mensagens_modelo = [('system', contexto)]
    mensagens_modelo += mensagens

    template = ChatPromptTemplate.from_messages(mensagens_modelo)
    chain = template | chat

    return chain.invoke({}).content

#Início do chat
contexto = definir_contexto(opcao)

print('\nBem-vindo ao Univille AI! (Digite x para sair)\n')

mensagens = []

while True:
    pergunta = input('Usuário: ')

    if pergunta.lower() == 'x':
        break

    mensagens.append(('user', pergunta))

    resposta = resposta_bot(mensagens, contexto)

    mensagens.append(('assistant', resposta))

    print(f'Bot: {resposta}\n')

#Histórico final
print('\nMuito obrigado por usar o Univille AI!')
print('Histórico da conversa:')
print(mensagens)
```
